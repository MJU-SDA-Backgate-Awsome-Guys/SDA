```python
import re

def list_food(target) :
    result = []
    target = re.sub("/", ' ',target).split(' ')
    for food in target.copy() :
        if '(' not in food and food != '' :
            result.append(food)
    return result
```

```python
menu = pd.DataFrame(data['중식메뉴'].apply(list_food))['중식메뉴']
menu_vector = menu.apply(str)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer().fit(menu_vector)

print("등장 하는 메뉴의 개수 : ",len(vect.vocabulary_.keys()),"\n\n 메뉴별 등장 횟수 : " ,vect.vocabulary_)
```

```python
bow = vect.transform(menu_vector)
```

```python
# 2016-02-01일 등장하는 메뉴의 개수는 8개, 따라서 1로 인코딩된 메뉴의 개수는 8개여야 한다.
pd.DataFrame(bow.toarray()[0]).value_counts()
```

```python
def split_for_label(x) :
    target = x.split(' ')
    result = []
    for food in target.copy() :
        if '(' not in food and food != '' :
            result.append(food)
    rice = result[0].split('/')[0].split('*')[0]
    soup = result[1]
    main = result[2]
    return [rice,soup,main]

    
menu_list = data['중식메뉴'].apply(split_for_label)
```

```python
rice = []
soup = []
main = []


for i in menu_list :
    rice.append(i[0])
    soup.append(i[1])
    main.append(i[2])
    
label_encoding = pd.DataFrame()
label_encoding['rice'] = rice
label_encoding['soup'] = soup
label_encoding['main'] = main
#label_encoding = pd.DataFrame(menu)['중식메뉴'].apply(three_meal)


label_encoding['rice'].value_counts()
#for x in label_encoding['soup'].value_counts().index :
#    print(x) #몇 몇 국,찌개가 정상적으로 저장되지 않았음을 확인
    
for i, soup in enumerate(label_encoding['soup']):
    if ("찌개" not in soup) and ("국" not in soup) and ("탕" not in soup) \
    and ("개장" not in soup) and ("시루" not in soup) and ("짬뽕" not in soup) and ("스프" not in soup) and\
    ("우동" not in soup) and ("수제비" not in soup) :
        print(i)
```
